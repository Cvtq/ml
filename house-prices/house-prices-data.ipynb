### Очистка от аномалий

```python
data2['Bedrooms']
```

Столбец Bedroom содержит как числовые, так и строковые значения

Следует это исправить

```python
from statistics import mean, pvariance, variance      # среднее арифметическое и дисперсии

# в столбце Bedrooms заменим все нечисловые значения на 0 
data2['Bedrooms'] = pd.to_numeric(data2['Bedrooms'], errors='coerce').fillna(0) #If 'coerce', then invalid parsing will be set as NaN.

# и в последствии удалим эти строки
mask = (data2['Bedrooms'] != 0)
data2 = data2[mask]

# заменим полученные ранее нули на среднее
var_val = mean(data2['Bedrooms'])
data2['Bedrooms'] = data2['Bedrooms'].replace(0, var_val)

data2['Bedrooms']
```

```python
from sklearn.preprocessing import LabelEncoder        # класс для кодирования признаков

# снова сделаем копию данных
# в новую таблицу далее запишем уже кодированные варианты признаков
data_enc = data2.copy()

label_encode_type = LabelEncoder()
# fit_transform -- просмотр всех значений в столбце, 
#                  создание таблицы соответствия: не_числовое_значение_признака - числовой_код
#                  преобразование всех значений в числа
data_enc['type_small'] = data2['Type'].apply( lambda s: s[:s.find(',')]) 
data_enc['type_small'] = label_encode_type.fit_transform( data_enc['type_small'] )

label_encode_state = LabelEncoder()
# fit_transform -- просмотр всех значений в столбце, 
#                  создание таблицы соответствия: не_числовое_значение_признака - числовой_код
#                  преобразование всех значений в числа
data_enc['State'] = label_encode_state.fit_transform( data2['State'] )

label_encode_type.classes_
```

#### Стандартизация

Признаки имеют разный масштаб. Математически это означает что одни признаки могут влиять на целевую переменную сильнее из-за большого разброса значений.

Однако фактически этот признак может не иметь сильного влияния на целевую переменную.

Для этого нужно привести все признаки к одному масштабу.

**Нормализация или стандартизация** -- приведение разных признаков к одной шкале, с одним и тем же масштабом.

Нормализация нужна чтобы уравнять относительную значимость признаков перед применением моделей машинного обучения. 
Например в линейной регрессии признаки с большим диапазоном значений, например 0-500 могут влиять на результат сильнее, чем признаки с диапазоном значений 1-80. Хотя важность признака должна определятся исследователем и моделью.

Нормализация vs стандартизация
- стандартизация $ X' = \frac{X - \bar X} {\sigma_X} $
    - после стандартизации: $\bar{X'} = 0$, &nbsp; $\sigma_{X'} = 1$
    - отлично подходит для нормально распределённых признаков
    - преобразование слабо зависит от выбросов
- нормализация $ X' = \frac{X-X_{min}}{X_{max} - X_{min}}$
    - после нормализации: $ X_{min}^{'} = 0$, &nbsp; $ X_{max}^{'} = 1$
    - подходит когда распределение признака неизвестно
    - зависит от выбросов


```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

#Стандартизация Тренировочной выборки
Train_SoldLess = Train.drop(columns='Sold Price')
Train_Sold = Train['Sold Price']
Train_SoldLess = scaler.fit_transform(Train_SoldLess)

#Стандартизация Тестовой выборки
Test_SoldLess = Test.drop(columns='Sold Price')
Test_Sold = Test['Sold Price']
Test_SoldLess = scaler.transform(Test_SoldLess)
```

```python
y_data = pd.concat([Train_Sold, Test_Sold])
y_data.to_csv("house-prices-y.csv")

x_data = pd.concat([pd.DataFrame(Test_SoldLess), pd.DataFrame(Train_SoldLess)])
x_data.to_csv("house-prices-x.csv")

train_df_x = pd.DataFrame(Train_SoldLess)
train_df_x.to_csv("train-house-prices-x.csv")

train_df_y = pd.DataFrame(Train_Sold)
train_df_y.to_csv("train-house-prices-y.csv")
```
